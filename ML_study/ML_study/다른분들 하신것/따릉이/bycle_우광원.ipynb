```python
import pandas as pd

ta_data = pd.read_csv(r'따릉이\train.csv')

te_data = pd.read_csv(r'따릉이\test.csv')

data = pd.concat([ta_data,te_data])

data
```

```python

data['count'][data['count'].isna()] = (-1-data.index)
data
```

```python
data[data['count']>0].iloc[:,-1]
```

```python
data = data.fillna(0)
```

```python
from sklearn.utils import all_estimators
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import warnings


# 경고 문구 모두 무시
warnings.filterwarnings('ignore')

# 데이터셋 준비
x, y = data[data['count']>0].iloc[:,:-1],data[data['count']>0].iloc[:,-1]
print(x.shape,y.shape)
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=1)

# 모델 로드
allAlgorithms = all_estimators('regressor')
# all_estimators 안에 있는 것들 중 'regressor'를 필터로 가져와서 쓴다.

# 모든 모델 돌면서 결과 출력
algo = {}
for (name, algorithm) in allAlgorithms:
    try:
        model = algorithm()
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
        algo[algorithm] = mean_squared_error(y_test, y_pred,squared=False)
        # print(name, '의 RMSE', algo[name])
    except:
        print(f'{name},{algorithm} - 모델 없음')
        continue
for i in sorted(algo.items(),key = lambda x: x[1]):
    print(i)

```
