#### 데이터 전처리

```python
# 불필요한 행 삭제

train = train.drop('id', axis = 1)
test = test.drop('id', axis = 1)
train.head()
```

```python
# 결측치가 8개 이상인 경우 삭제
train.dropna(thresh=8, inplace= True)
train.isnull().sum()
```

```python
#KNN imputer 이용해 결측치 대체

from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors = 5) 
train_x = imputer.fit_transform(train)
train_x = pd.DataFrame(train_x, columns=train.columns)

test_x = imputer.fit_transform(test)
test_x = pd.DataFrame(test_x, columns=test.columns)

train_x.isnull().sum(), test_x.isnull().sum()
```

```python
X_train_scaled=robustscaler.fit_transform(train_x)
X_test_scaled=robustscaler.fit_transform(test_x)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=test.columns)

# X_train_robust = robustscaler.transform(train_x) 
# test데이터 스케일링
# X_test_robust= robustscaler.transform(test_x)
```

```python
X_train_scaled = X_train_scaled.drop(['count'],axis=1)

# train_x=train_x.drop(['count'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 학습 데이터와 테스트 데이터 분리
# 타겟의 구성 비율에 맞게 학습용, 테스트용 검증용 테이터 셋 준비 (data, target random_state, strtify=target)

train_y=train['count']
X_train_scaled, val_x, train_y, val_y = train_test_split(X_train_scaled, train_y, test_size=0.25, random_state=42)
```

```python
rf = RandomForestRegressor(random_state=11)
rf.fit(X_train_scaled, train_y)
```

```python
y_pred = rf.predict(val_x)
# 예측 라벨과 실제 라벨 사이의 정확도 측정

mean_squared_error(y_pred, val_y)
r2_score(y_pred, val_y)

# 0.7012686805181497
```
