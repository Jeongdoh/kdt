###  스텝 함수 step()
- 반환 : 0 또는 1
- 기준(임계값) :  0

```python
def step(x):
    return np.array(x > 0, dtype=np.int32)
```

```python
x = np.arange(-5.0, 5.0, 0.1) # -5.0부터 5.0까지 0.1 간격 생성
y = step(x)
```

```python
features=[1.2, 3.0, 2.1, 4.1]
w = np.random.normal(size=4)
features, w + 0
```

```python
y=features * w + 0
y
```

```python
step(y)
```

```python
plt.title('Step Function')
plt.plot(x,y)
plt.show()
```

###  시그모이드 함수 : sigmoid()

- 임계값 : 0.5 
- 반환값 : 0.0 ~ 1.0

```python
# 시그모이드 함수 그래프를 그리는 코드
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
```

```python
np.argmax(y)
```

```python
plt.plot(x, y)
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

## 하이퍼블릭탄젠트 :  tanh()

- 임계값 : 0
- 값범위 : -1.0 ~ 1.0

```python
x = np.arange(-5.0, 5.0, 0.1) # -5.0부터 5.0까지 0.1 간격 생성
y = np.tanh(x)
```

```python
y
```

```python
plt.plot(x, y)
plt.plot([0,0],[1.0,-1.0], ':')
plt.axhline(y=0, color='orange', linestyle='--')
plt.title('Tanh Function')
plt.show()
```

## relu
- 반환값 : 0 ~ 
- 임계값 : 0
- 0보다 크면 입력값으로, 0보다 작으면 0
- f(x) = max(0, x)

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
```

```python
y
```

```python

plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Relu Function')
plt.show()
```

### Leaky_relu
- 입력값이 음수일 경우에 0이 아니라 0.001과 같은 매우 작은 수를 반환
- f(x) = max(ax, x)

```python
def leaky_relu(x):
    a=0.1
    return np.maximum(a*x, x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x)
```

```python
y
```

```python


plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Leaky ReLU Function')
plt.show()
```
