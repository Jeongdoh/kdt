```python
land=[]


for i in wiki['국가']:
    if i in asia: land.append('asia')

    elif i in erope: land.append('erope')

    elif i in oceania: land.append('oceania')

    elif i in america: land.append('america')

    elif i in africa: land.append('africa')

    else: land.append('other')
```

```python
wiki['land']=land
```
