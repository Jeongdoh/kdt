```python
# 결론

# 각 인구수 순위

# 북구>남구>동구>서구


# 1위 북부도서관 =정도 10월 1958권      /       대구 북구 지역의 어린이 인구가 굉장히 많은 것을 알 수 있음

# 2위 동부도서관 =도영 9월 1451권       /       인구수는 달서구가 많았지만 동부도서관이 독서량이 근소하게 높을 것을 알수있음(두류도서관과 40권 차이)

# 3위 두류도서관 =홍열 9월 1411권       /       동부도서관과 독서량이 차이가 근소하여 인구수와 어느정도 비례한다고 볼 수 있음

# 4위 서부도서관 =도형 9월 991권        /       서부도서관이 독서량, 인구수가 현저히 낮음


# 각 구의 인구수와 각 구의 시립도서관의 인기도서의 독서량이 어느정도 상관관계가 있다고 할 수 있음



# 각 구의 인구와 인기도서목록의 상관관계

# 인기도서목록을 인구와 비교하면 충분히 상관관계가 있다고 보여짐.
# 조사한 자료에 따르면 4개 구의 시립도서관 인기도서 목록의 공통 키워드 어린이 도서였으며
# 0~19세 연령의 인구가 많을 수록 각 구 시립도서관의 인기도서 1~10위까지의 어린이 도서를 많이 읽는다는것을 알 수 있음.
# 서구에는 각 구와 비교하였을때 어린이 인구가 가장 적고 로맨스 소설이 인기도서에 있는 만큼 인기도서 1~10위까지의 구독량과 상관관계가 있음.



# 만족도 상승을 위한 제안활성화 방법

# 출판사에서 책의 홍보활동을 한다면 *어린이 도서를 타겟*으로 해야하고 
# 4개의 구 중에서 / 북구, 달서구, 남구에서 하는게 가장 효율 적일 것 같고 각 구 시립도서관과 연계하여
# 어린이 도서 홍보를 위한 행사를 더욱 늘려가면 구독자의 만족도 상승으로 이어질 것으로 생각됨
# 서구에는 다른 구와 달리 로맨스 소설이나 인문학, 사회, 경제 분야등 성인이 읽을 만한 도서들에 대해 행사를 실시하고 더욱 다양한 분야의 스펙트럼을
# 제시하면 서구 시립도서관에서 타겟으로 생각하는 구독자들에게 큰 호응을 얻을 수 있을 것이라고 보여짐.
```
