```python
## 데이터 시각화==========================================================
# 1개 데이터 => y축으로 자동설정
plb.plot(oneDF.y, 'ro--')
```

```python
# 2개 데이터 => x, y축으로 자동설정
plb.plot(oneDF.x, oneDF.y, 'ro')
plb.xlabel('X')
plb.ylabel('Y')
plb.title("[anscombe]")
```

```python
# 2개 데이터 => x, y축으로 자동설정
plb.plot(oneDF.x, oneDF.y, 'ro', oneDF.x, oneDF.y*0.5, 'g^--')
plb.xlabel('X')
plb.ylabel('Y')
plb.title("[anscombe]")
```
