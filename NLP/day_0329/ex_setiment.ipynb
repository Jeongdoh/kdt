### [3] 데이터셋 전용 단어사전 생성

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# 데이터셋 사전 관련 상수
NUM_SIZE=10000
OOV='<UNK>'
```

```python
tokenizer=Tokenizer(num_words=NUM_SIZE,oov_token=OOV)
```

```python
tokenizer.fit_on_sequences(datas)
```

```python
# TEXT ==> Number 변환
tokenizer.texts_to_sequences(datas)
```
