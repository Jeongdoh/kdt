```python
from selenium import webdriver
import time

song20 = ['INVU', 'Candy Sugar Pop', 'ANTIFRAGILE', 'Lights']

driver = webdriver.Chrome('C:\\Program Files\\Google\\Chrome\\Application\\chromedriver.exe')
driver.get("https://www.lyrics.co.kr/")

lyrics = []
for song in song20:
    # 검색어 입력
    search_box = driver.find_element_by_id('gsc-i-id1')
    search_box.send_keys(song)
    search_box.submit()
    
    # 검색 결과에서 가사 페이지 링크 클릭
    try:
        link = driver.find_element_by_css_selector('.music_row > div > a')
        link.click()
    except:
        lyrics.append('No lyrics found for ' + song)
        driver.back()
        continue
    
    # 가사 페이지에서 가사 가져오기
    try:
        lyrics_box = driver.find_element_by_xpath('//*[@id="wrap"]/div[2]/div/div[2]/div/div[1]/pre')
        lyrics_text = lyrics_box.text
        lyrics.append(lyrics_text)
    except:
        lyrics.append('Failed to fetch lyrics for ' + song)
    
    driver.back()
    time.sleep(1)

driver.quit()

print(lyrics)
```
