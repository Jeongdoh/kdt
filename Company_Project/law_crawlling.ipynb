```python
import time
import random
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


times=random.randint(3,9)

# 크롬 드라이버 경로 설정
driver_path = 'C:\\Program Files\\Google\\Chrome\\Application\\chromedriver.exe'

# 국가법령정보센터 사이트 접속
driver = webdriver.Chrome(driver_path)
driver.get("https://www.law.go.kr/LSW/lsSc.do?section=&menuId=1&subMenuId=15&tabMenuId=81&eventGubun=060101&query=%EC%86%8C%EB%B0%A9%EA%B8%B0%EB%B3%B8%EB%B2%95#undefined")
time.sleep(2)
driver.implicitly_wait(10)

law1 = []
# 조문 추출
lyrics = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, 'bl')))
time.sleep(times)
for i in lyrics:
    law1.append(i.text)
# law.append(lyrics.text)

# 크롬 브라우저 종료
driver.quit()

# 리스트를 파일에 저장
with open('소방기본법.csv', 'w', encoding='utf-8') as f:
    for law in law1:
        f.write(law)
        f.write('\n')
```

```python
import re
import pandas as pd


with open('./소방시행규칙전체.txt', 'r', encoding='UTF-8') as file:
    text = file.read()



# 제목 및 내용 추출
pattern = re.compile('제(\d+)조\((.*?)\)(.*?)\n')
result = pattern.findall(text)

# 데이터 프레임 생성
columns = ['조문', '제목', '내용']
df2 = pd.DataFrame(result, columns=columns)

# 출력
print(df2)
```

```python
import csv
from konlpy.tag import Okt
from nltk.tokenize import word_tokenize
from nltk.metrics import jaccard_distance

# def read_csv(file_path):
#     data = []
#     with open(file_path, 'r', encoding='utf-8') as csv_file:
#         csv_reader = csv.reader(csv_file)
#         for row in csv_reader:
#             data.append(row)
#     return data

def preprocess_text(text):
    okt = Okt()
    tokens = okt.morphs(text)
    return tokens

def calculate_jaccard_similarity(tokens1, tokens2):
    set1 = set(tokens1)
    set2 = set(tokens2)
    jaccard_similarity = 1 - jaccard_distance(set1, set2)
    return jaccard_similarity

# # File paths
# file_path1 = './해양경비법전체.csv'
# file_path2 = './해양경비법시행령전체.csv'

# # Read CSV files
# data1 = read_csv(file_path1)
# data2 = read_csv(file_path2)

# Extract text from CSV data
texts1=[]
for i in range(df1.shape[0]):
    texts1.append(df1.iloc[i,1])

texts2=[]
for i in range(df2.shape[0]):
    texts2.append(df2.iloc[i,1])

# Tokenize and preprocess texts
preprocessed_texts1 = [preprocess_text(txt1) for txt1 in texts1]
preprocessed_texts2 = [preprocess_text(txt2) for txt2 in texts2]

# Calculate Jaccard similarity for each pair of texts
similarity_scores = []
for tokens1 in preprocessed_texts1:
    l = []
    for tokens2 in preprocessed_texts2:
        similarity = calculate_jaccard_similarity(tokens1, tokens2)
        l.append(similarity)
    similarity_scores.append(l)

# Print the similarity scores
for score in similarity_scores:
    print(score)
```
